Assignment: Majority Element (Easy)

Problem Given a list of integers nums, return the element that appears more than n/2 times.

You may assume a majority element always exists.

Example
nums = [3, 2, 3] → 3

nums = [2, 2, 1, 1, 1, 2, 2] → 2
Goal Solve using a dictionary that maps: number -> count.


```python
def majority_elements(nums):
  counts = {}

  for num in nums:
      if num in counts:
        counts[num]+= 1
      else:
        counts[num] = 1

print(majority_elements([3, 2, 3]))
```

```python
#Answer
def majority_elements(nums :list):
  counts={}

  for x in nums:
    counts[x]= counts.get(x,0) +1 #grabs the number from every item in the list and adds one to every times it's found
    #if key exists then grab the value and add 1 because we made the defaut 0 in case it was not there so it's zero and then adds one
    if counts[x] > len(nums)//2:
      return f"{x} appears the most"
  return -1
  #print(counts)

print(majority_elements([1,2,3,4,4,4,4,4]))
```
