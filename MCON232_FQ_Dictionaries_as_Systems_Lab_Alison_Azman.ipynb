### TODO 1 — add_student (answers above)
Create the student entry with an empty `grades` dict.

**Expected behavior**
- If the student already exists, do NOT overwrite their grades; either:
  - return False, or
  - raise a ValueError, or
  - politely do nothing and return False

Choose one behavior and be consistent.


````python
def add_student(school, name, age, major):
    # TODO: implement
  if not isinstance(school, dict):
    return False
  if not isinstance(name, str) or not isinstance(major, str):
    return False
  if not isinstance(age, int):
    return False
  if name in school:
    return False

  school[name] = {"age": age, "major" : major, "grades": {}}

  return True
````

### TODO 2 — add_grade
Add/overwrite the grade for a course inside `school[name]["grades"]`.

**Expected behavior**
- If the student does not exist, return False (or raise ValueError — your choice).
- Grades are ints 0–100 (you can validate if you want).


````python
def add_grade(school, name, course, grade):
    # TODO: implement
  if not isinstance(school, dict):
    return False
  if name not in school or not isinstance(school[name],dict):
    return False
  if "grades" not in school[name] or not isinstance(school[name]["grades"], dict):
    school[name]["grades"]={}
  if not isinstance(course, str):
    return False
  if not isinstance(grade, int) or not (0<=grade <=100):
    return False

  school[name]["grades"][course] = grade

  return True
````

### TODO 3 — get_student_average
Compute the average of the student’s grades.

**Expected behavior**
- If student missing: return None
- If student has no grades: return None


````python
def get_student_average(school, name):
    # TODO: implement
  if not isinstance(school, dict):
    return None
  if name not in school or not isinstance(school[name],dict):
    return None

  grades = school[name]["grades"]
  if not isinstance(grades, dict) or len(grades)==0:
    return None

  total=0
  count=0

  for grade in grades.values():
    if isinstance(grade, int):
      total+=grade
      count+=1
  if count==0:
    return None

  return (total/count)
````

Return a list of city codes (the dictionary keys like "NYC", "BOS") whose weather condition matches the condition string.

Example:

If condition == "sunny", return ["NYC", "MIA"] (whatever matches in your data)

Matching rule (simple + teachable):

compare case-insensitively (so "Sunny" and "sunny" both match)


````python
weather_api = {
    "NYC": {"temp": 42, "humidity": 70, "forecast": ["rain", "cloudy", "sun"]},
    "MIA": {"temp": 80, "humidity": 85, "forecast": ["sun", "sun", "sun"]},
    "DEN": {"temp": 35, "humidity": 40, "forecast": ["snow", "sun", "wind"]},
    "SEA": {"temp": 48, "humidity": 78, "forecast": ["rain", "rain", "cloudy"]},
}

weather_api
````

````python
def get_weather(api, city):
  if not isinstance(api, dict):
    return False
  # if not isinstance(city, dict) or not isinstance(api[city], dict):
  #   return False
  # if not isinstance(city["temp"], int):
  #   return False
  if city in api and isinstance(city,str):
    return api[city]["temp"]
  return {"error" : f"{city} not found"}

def average_temp(api):
    # TODO
  if not api or not isinstance(api, dict):
    return None

  total=0
  count=0

  for city_code, info in api.items():
    if not isinstance(info["temp"], (int, float)):
      return None
    total+=info["temp"]
    count+=1
  return (total/count)if count > 0 else None #comprehension

def cities_with_condition(api, condition):
    # TODO
   if not isinstance(api, dict):
       return []
   target = str(condition).strip().lower()


   matches = []
   for city_code, info in api.items():
       if not isinstance(info, dict):
           continue
       forecast_list = info.get("forecast", [])
       if any(str(x).lower() == target for x in forecast_list):
           matches.append(city_code)
   return matches
````
